# Polynomial Curve fitting of sine wave - 2 cycles

```python
import numpy as np
import matplotlib.pyplot as plt

# Create an array X of evenly spaced values
X = np.linspace(0, 2, 40)

# Generate Y using the sine function with Gaussian noise
np.random.seed(0)  # Set seed for reproducibility
noise = np.random.normal(0, 0.2, X.shape)
Y = np.sin(2 * np.pi * X) + noise
```

```python
plt.scatter(X,Y)
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import make_pipeline
```

```python
# Function to fit and plot polynomial curves for each degree individually
def plot_polynomial_fits(degrees, X, Y):
    X_reshaped = X[:, np.newaxis]  # Reshape X for model fitting

    for degree in degrees:
        # Create and fit the polynomial model
        model = make_pipeline(PolynomialFeatures(degree), LinearRegression())
        model.fit(X_reshaped, Y)
        Y_pred = model.predict(X_reshaped)

        # Plot the polynomial fit individually
        plt.scatter(X, Y, color='blue', label='Noisy Data')  # Scatter plot of noisy data
        plt.plot(X, Y_pred, label=f'Degree {degree}', color='red')

        # Plot settings for individual plot
        plt.title(f'Polynomial Curve Fitting (Degree {degree})')
        plt.xlabel('X')
        plt.ylabel('Y')
        plt.legend()
        plt.show()
```

```python
# Degrees to compare
degrees = [0,1, 2, 5,7,9,3,15,21]

# Call the function to plot polynomial fits for the given degrees
plot_polynomial_fits(degrees, X, Y)
```

```python
model = make_pipeline(PolynomialFeatures(7), LinearRegression())
model.fit(X[:, np.newaxis], Y)
```

```python
new_x = np.linspace(0, 2.2, 40)
new_y = model.predict(new_x[:, np.newaxis])
```

```python
plt.scatter(X, Y, color='blue', label='Noisy Data')
plt.plot(new_x, new_y, label='Polynomial Fit', color='red')
```

# Polynomial Curve fitting with sine wave - single cycle

```python
X = np.linspace(0, 1, 20)

# Generate Y using the sine function with Gaussian noise
np.random.seed(0)  # Set seed for reproducibility
noise = np.random.normal(0, 0.2, X.shape)
Y = np.sin(2 * np.pi * X) + noise
```

```python
# Degrees to compare
degrees = [0,1, 2,5,7,9,3,15,21]

# Call the function to plot polynomial fits for the given degrees
plot_polynomial_fits(degrees, X, Y)
```

```python
model = make_pipeline(PolynomialFeatures(3), LinearRegression())
model.fit(X[:, np.newaxis], Y)

new_x = np.linspace(0, 1.2, 20)
new_y = model.predict(new_x[:, np.newaxis])
plt.scatter(X, Y, color='blue', label='Noisy Data')
plt.plot(new_x, new_y, label='Polynomial Fit', color='red')
```
